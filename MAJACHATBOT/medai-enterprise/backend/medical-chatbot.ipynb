#  <b><span style='color:#15ef91 ; font-family:Goudy Stout'>Word Tokenization</span></b>

```python
import nltk
from nltk.tokenize import word_tokenize
def tokenize_text(text):
    tokens = word_tokenize(str(text))
    return tokens
for col in df.columns:
    df[col + '_tokens'] = df[col].apply(tokenize_text)
df.to_csv("tokenized_data.csv", index=False)
```

#  <b><span style='color:#15ef91 ; font-family:Goudy Stout'>Concatenate Dataset</span></b>

```python
data = pd.read_csv("tokenized_data.csv")
```

#  <b><span style='color:#15ef91 ; font-family:Goudy Stout'>Tranform into Lowercase</span></b>

```python
def lowercase_text(text):
    return str(text).lower()
df = data.applymap(lowercase_text)
```

#  <b><span style='color:#15ef91 ; font-family:Goudy Stout'>Removing Noise </span></b>

```python
import re

def remove_noise(text):
    text = re.sub(r'<.*?>', '', str(text))
    text = re.sub(r'[^\w\s]', '', text)
    return text
dff = df.applymap(remove_noise)
```

#  <b><span style='color:#15ef91 ; font-family:Goudy Stout'>Applying Stopwords </span></b>

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))
def remove_stopwords(text):
    words = text.split()  # Split text into words
    filtered_words = [word for word in words if word.lower() not in stop_words]  # Remove stopwords
    return ' '.join(filtered_words)  # Join filtered words back into a string

# Apply stopword removal to all columns
stdf = dff.applymap(remove_stopwords)
```

#  <b><span style='color:#15ef91 ; font-family:Goudy Stout'>Applying PorterStemmer </span></b>

```python
import nltk
from nltk.stem import PorterStemmer
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
stemmer = PorterStemmer()
def stem_text(text):
    words = nltk.word_tokenize(str(text))  # Tokenize text
    stemmed_words = [stemmer.stem(word) for word in words]  # Stem each word
    return ' '.join(stemmed_words)  # Join stemmed words back into a string
stmdf = stdf.applymap(stem_text)
```

#  <b><span style='color:#15ef91 ; font-family:Goudy Stout'>Applying TextBlob </span></b>

```python
from textblob import TextBlob
def spell_check_text(text):
    blob = TextBlob(str(text))
    corrected_text = blob.correct()
    return str(corrected_text)
sss = stmdf.applymap(spell_check_text)
```
